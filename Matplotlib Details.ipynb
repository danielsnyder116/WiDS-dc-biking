```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
plt.style.use('ggplot')
```

```python
#Get data to use
df = pd.read_csv('../quakers_nodelist.csv')
df['age'] = df['Deathdate'] - df['Birthdate']
```

```python
data = df.loc[0:5, ['Name','age']]
```

```python
data
```

```python
#Initializes figure - long way
fig = plt.figure(figsize=(10,4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.barh(data=data, y='Name', width='age')
ax1.set_title('Set Title Here')


ax2.bar(data=data, x='Name', height='age', color='tab:blue')
ax2.set_title('What?')
ax2.set_xlabel('People')
ax2.set_ylabel('Age')
ax2.set_xticks(data['Name'])

plt.show()
```

```python
#Short Way #Second Option to Above
fig, (ax1, ax2)= plt.subplots(1,2, figsize=(10,4))
fig,(ax1, ax2) = plt.subplots(2,2, figsize=(6,4))

plt.show()
```
