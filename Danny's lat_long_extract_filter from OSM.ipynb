```python
import pandas as pd
import numpy as np
import sys
import os
import xml
from xml.dom.minidom import parse
```

```python
# name of file
xmls = [r"district-of-columbia-latest.osm"]
```

```python
# create dictionary to store data
data = {}
```

```python
# loop through file, find all "node" tags and extract all "lat" and "lon" attributes
for f in xmls:
    dom = xml.dom.minidom.parse(f)
    for n in dom.getElementsByTagName("node"):
        nid = n.getAttribute("id")
        data[nid] = {}
        data[nid]["lat"] = n.getAttribute("lat")
        data[nid]["lon"] = n.getAttribute("lon")
```

```python
data
```

```python
# pull out "lat" and "lon" from data dictionary
latitude_string = [data[key]['lat'] for key in data]
longitude_string = [data[key]['lon'] for key in data]
```

```python
# strip to only 4 decimal points
latitude_string_short = [lat[:7] for lat in latitude_string]
```

```python
# map variables from type string to float
latitude_float = list(map(float, latitude_string))
longitude_float = list(map(float, longitude_string))
```

```python
# zip points together to maintain coordinate pair structure
points = list(zip(latitude_float, longitude_float))
```

```python
# example of extracting lat and lon from points object
[item[0] for item in points[:5]] # for latitude
[item[1] for item in points[:5]] # for long
```

```python
# bottom left corner @ H street and 18th street NW
# top right corner @ L street and 14th street  NW
latitude_bounds = (38.900217, 38.903740)
longitude_bounds = (-77.041673, -77.031947)
```

```python
# yield a count for the number of points in the filtered square
# estimate of number of points, ballpark to check filtering works
count = 0
for point in points:
    if ((point[0] >= latitude_bounds[0] and point[0] <= latitude_bounds[1]) and \
            (point[1] >= longitude_bounds[0] and point[1] <= longitude_bounds[1])):
        count += 1
```

```python
# filter the points based on the latitude and longitude bounds from above
filtered_points = [point for point in points \
                       if ((point[0] >= latitude_bounds[0] and point[0] <= latitude_bounds[1]) and \
                           (point[1] >= longitude_bounds[0] and point[1] <= longitude_bounds[1]))]
```
