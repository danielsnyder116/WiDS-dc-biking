```python
# Assuming we are using Djikstra's algorithm, this will be the best option for weighting:

# Set it up so all of the edges start with a certain weighting (20):
# If the edge has a positive/safety aspect, we will subtract that value from the base weight
# If the edge has a negative/danger aspect, we will add that value to the base weight

# Once this is complete, we will have the safest routes as the least costly or lowest values and
# the most dangerous routes as the most costly or highest values
```

```python
# Adding the weights to each segid:
df_start['total_weight'] = 20
```

```python
#Takes a little while to run

# This way seems time consuming, maybe use the better way that Kevin showed you?
# Perhaps in this case this is actually the best option, as lambda functions are much better for filtering or completing
# one action on every row than determining the value of one column which determines how another column is changed.


#So go back and write a function that takes care of this, then put it into the .apply(lambda function) as it should run faster
# and at least will be cleaner


# NEED to aggregate by seg-id FIRST

#Initialize weight
for i in range(0, len(df_start)):
    
    ########## Less Safe Weights ############
    
    #Vision Zero Incident
    if df_start['REQUESTTYPE'][i] != 0:
        df_start['total_weight'][i] = df_start['total_weight'][i] + 1
        
    # Bicyclist Killed
    if df_start['FATAL_BICYCLIST'][i] > 0:
        df_start['total_weight'][i] = df_start['total_weight'][i] + 5* df_start['FATAL_BICYCLIST'][i]
    
    #Major Injury
    if df_start['MAJORINJURIES_BICYCLIST'][i] > 0:
        df_start['total_weight'][i] =  df_start['total_weight'][i] + 4* df_start['MAJORINJURIES_BICYCLIST'][i]
    
    #Minor Injury
    if df_start['MINORINJURIES_BICYCLIST'][i] > 0: 
        df_start['total_weight'][i] = df_start['total_weight'][i] + 3* df_start['MINORINJURIES_BICYCLIST'][i]
        
    #Bicyclist Impaired
    if df_start['BICYCLISTSIMPAIRED'][i] > 0:
        df_start['total_weight'][i] = df_start['total_weight'][i] + 3* df_start['BICYCLISTSIMPAIRED'][i]
        
    #Pothole   
    if df_start['POTHOLES_ADDRESSES'][i] != 0:
        df_start['total_weight'][i] = df_start['total_weight'][i] + 1
        
    #Lights per measure
    if df_start['NUMBER_LIGHTS'][i] / df_start['ENDMEASURE'][i] < .020:
         df_start['total_weight'][i] = df_start['total_weight'][i] + 1
            
    #Shared lane OR Shared Bus/Bike Lane
    if (df_start['FACILITY'][i] == 'Shared Lane') | (df_start['FACILITY'][i] == 'Bus/Bike Lane'):
        df_start['total_weight'][i] = df_start['total_weight'][i] + 1
            
    #--------------------------------------------------------------------#
    
    ############## Safer Weights #############
    
    #Presence of some sort of bike lane
    # NEED to figure out what to do for NaN values?
    
    #Existing Bike Line OR #Cycle Track OR #Climbing Lane OR #Contraflow Bike Lane
    if (df_start['FACILITY'][i] != 'Shared Lane') | (df_start['FACILITY'][i] != 'Bus/Bike Lane'):
        df_start['total_weight'][i] = df_start['total_weight'][i] - 1
            
```
